```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('inflation.csv')
```

```python
df.head()
```

```python
# Visualization 1: Time Series Plot of CPI
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Rate', data=df)
plt.title('Consumer Price Index (CPI) Over Time')
plt.xlabel('Year')
plt.ylabel('CPI')
plt.show()
```

```python
# Visualization 2: Month-over-Month Inflation Rate
plt.figure(figsize=(12, 6))
df['MoM_Inflation'] = df['Rate'].pct_change() * 100
sns.lineplot(x='Year', y='MoM_Inflation', data=df)
plt.title('Month-over-Month Inflation Rate')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.show()
```

```python
value = df['State'].value_counts()
final_value = value[0:5]
final_value
figure = plt.figure(figsize = (9,9), dpi = 150)
Labels = ['Texas', 'Virginia', 'Missouri', 'Kentucky', 'Kanas']
plt.bar(Labels, final_value)
plt.title('Top 5 States with highest number of unemployment')
plt.xlabel('States of USA')
plt.ylabel('Number of unemployment')
plt.show()
```

```python
values_of_county = df['County'].value_counts()
final_county = values_of_county[0:5]
final_county
figure = plt.figure(figsize = (9,9), dpi = 150)
pie_labels = ['Washington', 'Jefferson', 'Lincoln', 'Franklin', 'Jackson']
explode = [0.2,0,0,0,0]
plt.pie(final_county, labels = pie_labels, autopct = '%.1f %%', explode = explode,
       shadow = True, pctdistance = 0.9)
plt.title('Top 5 County with highest number of unemployment')
plt.legend()
plt.show()
```
